# Análise da Curva de Carga de Energia Elétrica — Brasil (2025)

Este notebook apresenta uma análise exploratória dos dados de curva de carga
de energia elétrica do Sistema Interligado Nacional (SIN), com foco em padrões
horários, comportamento por subsistema e identificação de picos de consumo.
M
#**Fonte dos dados:** Operador Nacional do Sistema Elétrico (ONS)

## Objetivos da Análise

- Entender a estrutura e granularidade da base
- Validar colunas e tipos de dados
- Analisar o comportamento da carga ao longo do tempo
- Criar base para visualizações e insights de negócio

## Importação das bibliotecas

```python
import pandas as pd
import numpy as np
```

## Leitura do dataset

```python
df = pd.read_csv(
    "../data/raw/CURVA_CARGA_2025.csv",
    sep=";",
    encoding="latin1"
)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

## Tratamento e preparação dos dados

```python
df["din_instante"] = pd.to_datetime(df["din_instante"])
```

```python
df.dtypes
```

```python
df = df.rename(columns={
    "id_subsistema": "subsistema_id",
    "nom_subsistema": "subsistema",
    "din_instante": "data_hora",
    "val_cargaenergiahomwmed": "carga_mw_media"
})
```

```python
df.columns
```

```python
df["hora"] = df["data_hora"].dt.hour
df["dia"] = df["data_hora"].dt.day
df["mes"] = df["data_hora"].dt.month
df["dia_semana"] = df["data_hora"].dt.day_name()
```

```python
df.head()
```

## Perguntas de Negócio

1. Qual é o padrão médio de consumo ao longo do dia?
2. Quais subsistemas concentram maior carga média?
3. Em quais horários ocorrem os picos de consumo?
4. Há diferença de comportamento entre dias da semana?

```python
curva_media_hora = (
    df
    .groupby("hora", as_index=False)
    .agg(carga_media_mw=("carga_mw_media", "mean"))
)

curva_media_hora.head()
```

```python
carga_por_subsistema = (
    df
    .groupby("subsistema", as_index=False)
    .agg(carga_media_mw=("carga_mw_media", "mean"))
    .sort_values("carga_media_mw", ascending=False)
)

carga_por_subsistema
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))
plt.plot(curva_media_hora["hora"], curva_media_hora["carga_media_mw"])
plt.xlabel("Hora do dia")
plt.ylabel("Carga média (MW)")
plt.title("Curva Média Diária de Carga — Brasil (2025)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(carga_por_subsistema["subsistema"], carga_por_subsistema["carga_media_mw"])
plt.xlabel("Subsistema")
plt.ylabel("Carga média (MW)")
plt.title("Carga Média por Subsistema — Brasil (2025)")
plt.show()
```

```python
ordem_dias = [
    "Monday", "Tuesday", "Wednesday",
    "Thursday", "Friday", "Saturday", "Sunday"
]

df["dia_semana"] = pd.Categorical(
    df["dia_semana"],
    categories=ordem_dias,
    ordered=True
)
```

```python
heatmap_data = (
    df
    .groupby(["dia_semana", "hora"])
    .agg(carga_media_mw=("carga_mw_media", "mean"))
    .reset_index()
)
```

```python
heatmap_pivot = heatmap_data.pivot(
    index="dia_semana",
    columns="hora",
    values="carga_media_mw"
)

heatmap_pivot.head()
```

```python
plt.figure(figsize=(14, 6))
plt.imshow(heatmap_pivot, aspect="auto")
plt.colorbar(label="Carga média (MW)")
plt.xticks(range(24), range(24))
plt.yticks(range(len(heatmap_pivot.index)), heatmap_pivot.index)
plt.xlabel("Hora do dia")
plt.ylabel("Dia da semana")
plt.title("Heatmap da Carga Média — Hora x Dia da Semana (2025)")
plt.show()
```

## Principais Insights

- O consumo apresenta picos recorrentes no período da tarde/noite,
  especialmente entre 18h e 21h.
- Dias úteis apresentam carga média superior ao fim de semana,
  com padrão mais estável ao longo do dia.
- Aos finais de semana, observa-se redução da carga e maior
  dispersão do consumo ao longo do dia.

```python
curva_subsistema = (
    df
    .groupby(["subsistema", "hora"], as_index=False)
    .agg(carga_media_mw=("carga_mw_media", "mean"))
)

curva_subsistema.head()
```

```python
plt.figure(figsize=(12, 6))

for subsistema in curva_subsistema["subsistema"].unique():
    dados = curva_subsistema[curva_subsistema["subsistema"] == subsistema]
    plt.plot(dados["hora"], dados["carga_media_mw"], label=subsistema)

plt.xlabel("Hora do dia")
plt.ylabel("Carga média (MW)")
plt.title("Curva Média Diária de Carga por Subsistema — Brasil (2025)")
plt.legend()
plt.grid(True)
plt.show()
```

## Insights por Subsistema

- O subsistema SE/CO concentra a maior carga média ao longo de todo o dia,
  apresentando picos mais acentuados no período noturno.
- Os subsistemas Norte e Nordeste apresentam perfis mais estáveis,
  com menor amplitude entre carga mínima e máxima.
- O subsistema Sul apresenta comportamento intermediário,
  com picos menos pronunciados em relação ao SE/CO.

```python
plt.figure(figsize=(10, 5))
plt.plot(curva_media_hora["hora"], curva_media_hora["carga_media_mw"])
plt.xlabel("Hora do dia")
plt.ylabel("Carga média (MW)")
plt.title("Curva Média Diária de Carga — Brasil (2025)")
plt.grid(True)

plt.savefig("../images/curva_media_diaria_brasil.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(carga_por_subsistema["subsistema"], carga_por_subsistema["carga_media_mw"])
plt.xlabel("Subsistema")
plt.ylabel("Carga média (MW)")
plt.title("Carga Média por Subsistema — Brasil (2025)")

plt.savefig("../images/carga_media_por_subsistema.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(12, 6))

for subsistema in curva_subsistema["subsistema"].unique():
    dados = curva_subsistema[curva_subsistema["subsistema"] == subsistema]
    plt.plot(dados["hora"], dados["carga_media_mw"], label=subsistema)

plt.xlabel("Hora do dia")
plt.ylabel("Carga média (MW)")
plt.title("Curva Média Diária por Subsistema — Brasil (2025)")
plt.legend()
plt.grid(True)

plt.savefig("../images/curva_diaria_por_subsistema.png", dpi=150, bbox_inches="tight")
plt.show()
```
